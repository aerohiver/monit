```python
# 랭체인 각 모듈 생성
llm = ChatOpenAI(openai_api_key=OPENAI_API_KEY)

prompt = ChatPromptTemplate.from_messages([
    ("system", "You must summarize the given text in 5 sentences or less in Korean."),
    ("user", "{input}")
])

output_parser = StrOutputParser()

# 체인화
chain = prompt | llm | output_parser
```

# 엑셀 서식 정리

```python
# 예를 들어, 첫 번째 행의 셀 배경색이 노란색, 글자 크기가 12, 셀 너비가 15, 가운데 정렬이라고 가정
width_list = [10, 20, 25, 25, 25, 70, 18]
```

```python
from openpyxl import load_workbook
from openpyxl.styles import PatternFill, Font, Alignment

# 파일 로드
wb = load_workbook("df_concat_with_llm.xlsx")
sheet = wb.active  # 활성 시트 선택

# 불필요한 1~2열 삭제
sheet.delete_cols(idx=1)  # A열의 인덱스는 1
sheet.delete_cols(idx=1)  # A열의 인덱스는 1

# ========================================
# 정렬
# ========================================

# A열에서 내용이 있는 마지막 행 찾기
max_row = sheet.max_row
for i in range(max_row, 0, -1):
    if sheet[f'A{i}'].value is not None:
        max_row_with_content = i
        break

# A,B,G열의 내용이 있는 행까지만 수평/수직 가운데정렬 적용, 나머지는 수직 가운데정렬만 적용

for row in sheet.iter_rows(min_row=1, max_row=max_row_with_content, min_col=1, max_col=2):
    for cell in row:
        cell.alignment = Alignment(horizontal='center', vertical='center')

for row in sheet.iter_rows(min_row=1, max_row=max_row_with_content, min_col=7, max_col=7):
    for cell in row:
        cell.alignment = Alignment(horizontal='center', vertical='center')

for row in sheet.iter_rows(min_row=1, max_row=max_row_with_content, min_col=3, max_col=6):
    for cell in row:
        cell.alignment = Alignment(vertical='center', wrapText=True)

# 헤드는 특별 관리
for cell in sheet[1]:  
    cell.font = font
    cell.alignment = Alignment(horizontal='center', vertical='center')
    
# ========================================
# 행높이, 열너비 조정
# ========================================     
sheet.row_dimensions[1].height = 31.2
for row in range(2, sheet.max_row + 1):  # ws.max_row는 마지막 행 번호
    sheet.row_dimensions[row].height = 120

# 셀 너비 조정
for col, width in zip(sheet.columns, width_list):
    sheet.column_dimensions[col[0].column_letter].width = width

    
# ========================================
# 헤드: 배경색, 폰트색 변경
# ========================================
fill = PatternFill(start_color='000000', end_color='000000', fill_type='solid') # 검정색
font = Font(color="F2F2F2", bold=True)

for cell in sheet[1]:  # 첫 번째 행의 모든 셀에 적용
    cell.font = font
    cell.fill = fill
    
# 변경 사항 저장
wb.save("df_concat_with_llm_a.xlsx")
```
